```
%matplotlib inline
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
%cd N:\john\Tool\realcloud\realcloud\user_uploads\wooey_scripts
F = open("table.csv")
d = {}
content = F.readlines()
l_header = content[0].rstrip().split("\t")
for i in l_header:
    d[i] = []
for line in content[1:]:
    fields = line.rstrip().split("\t")
    d[l_header[0]].append(float(fields[0]))
    d[l_header[1]].append(fields[1])
    d[l_header[2]].append(fields[2])

tips = pd.DataFrame(d)
sns.set(style="ticks", palette="pastel", color_codes=True)
plt.subplots(figsize=(12, 6))
dx = sns.violinplot(y="tax", x="Relative abundance", hue="Group", scale="width", data=tips, split=True, inner="quart", palette={"Health": sns.xkcd_rgb["peach"], "Carries": sns.xkcd_rgb["baby blue"]})
sns.despine()
plt.savefig("vio.png")
```
